```python
contents = [
    "Give me profiles of people working in export business",
    "Give me profiles of people who would be interested in reading articles about stock market ",
    "Give me profiles of people who would be using data analytics tools",
    "Give me profiles of people working who will be interested in consultancy roles",
    "Give me profiles of people who will be interested in attending finance workshops"
]
documents = [
    {"user_id" : 1, "content": content, "content_vector": embedder.encode(content, convert_to_numpy=True).tolist(), "response": ""} for content in contents
]
```

```python
from typing import Any
import requests
from elasticsearch import helpers

def send_to_elasticsearch(client, documents: dict[str, Any], index: str) -> Any:
    response = helpers.bulk(
        client,
        documents,
        index=index
    )
    return response

def vector_search(client, content_vector: list[float]) -> dict:
    script_query = {
        "script_score": {
            "query": {"match_all": {}},
            "script": {
                "source": "cosineSimilarity(params.query_vector, doc['content_vector'])",
                "params": {"query_vector": content_vector }
            }
        }
    }

    script_query = {
        "knn":{
            "k": 10,
            "num_candidates": 10,
            "field": "lead_ind",
            "query_vector": content_vector
        }
    }

    response = helpers.scan(
        client,
        index="leads_ind",
        body={
            "size": 10,
            "query": script_query,
            "_source": {
                "includes": ["content"]
            }
        }
    )

    response = client.knn_search(
        body=script_query,
        index="leads_ind"
    )["hits"]["hits"]

    return response

def find_document_by_id(client, doc_id: Any) -> dict:
    return helpers.scan(
        client,
        index="leads_ind",
        query={
            "query": {
                "match": {
                    "_id": doc_id
                }
            }
        }
    )

search_document = documents[2]


print(f"Document will find relevant queries for: \n\n{search_document['content']}\n\n")


search_results = vector_search(client, content_vector=search_document["content_vector"])

search_results = list(search_results)

for i in search_results:
    print(i)

if len(search_results) == 0:
    send_to_elasticsearch(client, documents, index="leads_ind")
if len(search_results) > 0:
    print("Search results = ", json.dumps({"similarity_score": search_results[0]["_score"], "question": search_results[0]["_source"]["content"]}, indent=4))
    required_doc = list(find_document_by_id(client, search_results[0]["_id"]))[0]
    for i in required_doc:
        print(json.dumps(i["_source"]["content"], indent=4))
    print("\n\n", json.dumps(required_doc["_source"]["content"], indent=4))
```
